```python
# lets put all code together and extract the titles,details,prices and review for 20 pages
import bs4
from bs4 import BeautifulSoup as bs
import requests
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup

title=[]
prices=[]
product_description = []
review_count = []
for i in range(1,20):
    link='https://www.officesupply.com/office-supplies/writing-correction/pens-pencils/c200236.html?p='+str(i)
    print(i)
    page = requests.get(link)
    soup = bs(page.content)

    soup = BeautifulSoup(html, "html.parser") 
    spans=soup.find_all('div', {'class': 'title'})
    titles = []
    for span in spans:
        titles.append(span.text.strip())

    for i in titles:
        if len(i) > 70:
            title.append(i)

    soup = BeautifulSoup(html, "html.parser") 
    spans=soup.find_all('div', {'class': 'short-description'})
    
    for span in spans:
        product_description.append(span.text.strip()) 

    soup = BeautifulSoup(html, "html.parser") 
    spans=soup.find_all('div', {'class': 'price'})
    price = []
    for span in spans:
        price.append(span.text.strip())

    for i in price:
        prices.append(i[0:6].replace("\n", ""))

    soup = BeautifulSoup(html, "html.parser") 
    spans=soup.find_all('span', {'class': 'review-count'})
  
    for span in spans:
        review_count.append(span.text.strip())
```

```python
import pandas as pd
df=pd.DataFrame({'Product Name':title,
                 'Price':prices,
                 'product_description':product_description,
                 'review_count':review_count
                })
df.head(60)
```
