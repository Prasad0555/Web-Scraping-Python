```python
url = 'https://www.appfutura.com/web-developers'
reqs = requests.get(url)
soup = BeautifulSoup(reqs.text, 'html.parser')
 
urls = []
for link in soup.find_all('a'):
    urls.append(link.get('href'))
    
for i in urls:
    if len(i) < 10:
        urls.remove(i)
```

```python
urls
```

```python
titles = []
for i in urls:
    spans=soup.find_all('a', {'href': i})
    for span in spans:
        titles.append(span.text.strip())
        print(span.text.replace('USD','').strip())
```
